# Context and Project Objectives

The aim of the following notebook is to explain the context of the current project and the questions to answer.

## Context
---
### Business context
The dataset corresponds to a UK-based online retail company with no physical stores, primarily selling unique all-occasion giftware. A significant portion of the company’s customers are wholesalers, which results in varying purchasing patterns and order volumes.
### Analytical role
In this project, I assume the role of a Junior Data Analyst supporting the Marketing and CRM teams. The objective is to analyze customer purchasing behavior and provide data-driven insights to improve customer retention and marketing strategies.
### Dataset
The Online Retail II dataset contains transactional data from December 1st, 2009 to December 19th, 2011, including information on invoices, products, quantities, prices, customers, and transaction dates.

## Project Objectives
---
### Main objective
The main objective of this project is to analyze customer purchasing behavior and segment customers using RFM analysis and clustering techniques, in order to support marketing and customer retention decision-making.

### Business questions
- How are customers distributed according to their recency, frequency, and monetary value?

- What are the main characteristics of each customer segment?

- What percentage of the company’s total revenue is generated by the most valuable customers?

- Are there customers with high purchase frequency but relatively low monetary value?

- What actions could the company implement to:

    - retain high-value customers?

    - reactivate inactive customers?

    - increase the spending of frequent but low monetary value customers?
